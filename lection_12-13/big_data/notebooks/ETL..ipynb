```scala
%spark.pyspark from pyspark.sql import functions as F

df = spark.read.csv("s3a://apache/ETL/", header=True)

df.show(10)
```

```scala
%spark.pyspark

df = df.filter(F.col("region") == "Western Europe")

df.show(10)
```

```scala
%spark.pyspark

df.write \
    .mode("overwrite") \
    .parquet("s3a://apache/ETL/world_happiness_transformed/")
```
